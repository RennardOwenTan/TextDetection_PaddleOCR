```python
import cv2
import keras_ocr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imread
from google.colab.patches import cv2_imshow
from paddleocr import PaddleOCR, draw_ocr
```

```python
ocr = PaddleOCR(use_angle_cls=True, lang="en")
```

```python
# Image Processing
img = cv2.imread("spiderman.png")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#gamma = np.array(255 * (img / 255) ** 3, dtype='uint8')

_, binarized = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

inverted = cv2.bitwise_not(binarized)

kernel = np.ones((1, 1), np.uint8)
eroded = cv2.erode(inverted, kernel, iterations=1)
#dilated = cv2.dilate(inverted, kernel, iterations=1)

test = eroded
cv2_imshow(test)
```

```python
img_ori = "spiderman.png"
result = ocr.ocr(test, cls=True)

data = []
for line in result[0]:
    coordinates = line[0]
    detected_text = line[1][0]
    confidence = line[1][1]

    data.append({
        "bbox": coordinates,
        "text": detected_text,
        "conf": confidence
    })
df = pd.DataFrame(data)

result_plot = df[['text', 'bbox']].values.tolist()
result_plot = [(x[0], np.array(x[1])) for x in result_plot]
keras_ocr.tools.drawAnnotations(plt.imread(img_ori), result_plot)
```

# App version

```python
import cv2
import numpy as np
import pandas as pd
from paddleocr import PaddleOCR
import matplotlib.pyplot as plt
import keras_ocr

img_path = None

ocr = PaddleOCR(use_angle_cls=True, lang='en')

def input_image():
    global img_path
    img_path = input("Enter the path to your image: ")
    try:
        img = cv2.imread(img_path)
        if img is None:
            print("Error: Unable to load the image. Please check the path.")
            img_path = None
        else:
            print("Image loaded successfully!")
    except Exception as e:
        print(f"Error: {e}")
        img_path = None

def detect_text():
    global img_path
    if not img_path:
        print("No image loaded. Please input an image first.")
        return

    try:
        # Image Processing
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        _, binarized = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)
        inverted = cv2.bitwise_not(binarized)
        kernel = np.ones((10, 3), np.uint8)
        eroded = cv2.erode(inverted, kernel, iterations=1)

        # Reader
        result = ocr.ocr(eroded, cls=True)
        data = []
        for line in result[0]:
            coordinates = line[0]
            detected_text = line[1][0]
            confidence = line[1][1]

            data.append({
                "bbox": coordinates,
                "text": detected_text,
                "conf": confidence
            })
        df = pd.DataFrame(data)
        result_plot = df[['text', 'bbox']].values.tolist()
        result_plot = [(x[0], np.array(x[1])) for x in result_plot]
        keras_ocr.tools.drawAnnotations(plt.imread(img_path), result_plot)
        plt.show()
        return

    except Exception as e:
        print(f"Error during text detection: {e}")

if __name__ == "__main__":
    while True:
        print("\n>> Stylized Text Detection ft. PaddleOCR <<")
        print("1. Input image")
        print("2. Detect text")
        print("3. Exit")

        try:
            choice = int(input(">> "))
            if choice == 1:
                input_image()
            elif choice == 2:
                detect_text()
            elif choice == 3:
                print("Thank you for using our app!")
                break
            else:
                print("Invalid choice!")
        except ValueError:
            print("Please enter a valid number.")
```
